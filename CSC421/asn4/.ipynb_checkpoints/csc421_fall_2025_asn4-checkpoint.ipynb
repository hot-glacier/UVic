 # Question 1 (Basic) - Generate HMM samples - 6 Marks


Explore hidden Markov models (HMMs) using the hmmlearn Python library:
[https://hmmlearn.readthedocs.io/en/latest/index.html]. 

You can use the code for the weather example in the probabilistic reasoning over time notebook we covered in class as a template/starting point for writing your code. 

The problem here is inspired by the use of HMMs in bioinformatics. There are several simplifications made to make it reasonable as part of an assignment. DNA sequences can be considered strings over an alphabet of 4 symbols/nucleobases **A,C,T,G (adenine, cytosine, thymine, guanine**. Parts of a DNA sequence are dense with C and G and other parts are sparse with C and G and it is of interest to biologists to identify these regions. 

We will model the **CG-dense**(CGD) and **CG-sparse** (CGS) as hidden states, and the nucleobases will be the observations. Through experimental data, we have the following information: 

1. The transition probability from CGD to CGS is 0.37 and the probability of staying in CGD is 0.63. The transition probability from CGS to CGD is similarly 0.37 with 0.63 being the probability of staying in CGS. 

2. The observation probabilities of CGD regions are A: 0.15, C:0.35, G: 0.35, and T:0.15. The observation probabilities of CGS regions are A: 0.40, C: 0.10, G: 0.10, T: 0.40 

3. You can assume that the initial state probabilities are the same (0.5) 

4. For visualization of the DNA sequences use the following color mapping: A: red, C: green, T: blue, G: yellow, and for CGD: black 
and CGS: white 


Your tasks:
1. Define this HMM model using the **hmmlearn** conventions.
2. Then use the created model to generate a sequence of 1000 samples (i.e. both hidden states and corresponding observations).
3. Use the colours above to visualize the sequence of samples. 



```python
# Your answer goes here 
%matplotlib inline 
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from hmmlearn import hmm

                    # CGD, CGS
transmat = np.array([[0.63, 0.37], #<- CGD
                    [0.37, 0.63]])  #<- CGS

start_prob = np.array([0.5, 0.5])

# green (C) and yellow (G) have high probs for CGD
# red (A) and blue (T) have high probs for CGS 
emission_probs = np.array([[0.15, 0.35, 0.35, 0.15], 
                           [0.40, 0.10, 0.10, 0.40]])

model = hmm.MultinomialHMM(n_components=2, n_trials=1)
model.startprob_ = start_prob 
model.transmat_ = transmat 
model.emissionprob_ = emission_probs

# sample the model - X is the observed values 
# and Z is the "hidden" states 
X, Z = model.sample(1000)

state2color = {} 
state2color[0] = 'black'
state2color[1] = 'white'

def plot_weather_samples(samples, state2color, title): 
    colors = [state2color[x] for x in samples]
    x = np.arange(0, len(colors))
    y = np.ones(len(colors))
    plt.figure(figsize=(10,1))
    plt.bar(x, y, color=colors, width=1)
    plt.title(title)
    
plot_weather_samples(Z, state2color, 'States')

samples = [entry.tolist().index(1) for entry in X]
obj2color = ['red', 'green', 'blue', 'yellow']

plot_weather_samples(samples, obj2color, 'Observations')
```

There is a company that sells boxes of pre-flipped coins. The pre-flipped coins are either heads or tails. These boxes come in 5 different variations:  

h1: 100% heads  
h2: 75% heads, 25% tails  
h3: 50% heads, 50% tails  
h4: 25% heads, 75% tails  
h5: 100% tails  

Given a new box of pre-flipped coins, the random variable H takes one of these 5 values: h1, h2, h3, h4, h5. We don't know which hypothesis it is and we gradually check coins D1, D2, D3,..., DN, where each of those is a random variable with value heads and tails.  

The task is given a sequence of observations D1, ... DN to predict the face of the next coin.  

The sellers of said pre-flipped coins have provided a prior probability distribution of the hypotheses. Naturally, as a tails result is more rare (must be why I lose coin flips a lot), the distribution is 10% h1, 10% h2, 20% h3, 30% h4, and 30% h5.  




```python
# The code portion of your answer goes here
import collections 
import numpy as np
from bokeh.plotting import figure, show 
from bokeh.io import output_notebook 

data_book1 = ['h'] * 10
data_book2 = ['h','t','h','h','h','h','h','t','h','t']
data_book3 = ['h','t','h','t','t','h','t','h','t','t']
data_book4 = ['t','t','t','t','t','t','t','h','h','t']
data_book5 = ['t'] * 10

counts_book = collections.Counter(data_book1)
print(counts_book['h'], counts_book['t'])
counts_other = collections.Counter(data_book2)
print(counts_other['h'], counts_other['t'])
counts_other = collections.Counter(data_book3)
print(counts_other['h'], counts_other['t'])
counts_other = collections.Counter(data_book4)
print(counts_other['h'], counts_other['t'])
counts_other = collections.Counter(data_book5)
print(counts_other['h'], counts_other['t'])

h1 = [1.0, 0.0]   # h1: 100% heads
h2 = [0.75, 0.25] # h2: 75% heads and 25% tails 
h3 = [0.5, 0.5]   # h3: 50% heads and 50% tails
h4 = [0.25, 0.75] # h4: 25% heads and 75% tails 
h5 = [0.0, 1.0]   # h5: 100% tails 

prior = [0.1, 0.1, 0.2, 0.3, 0.3]

def likelihood(d, h): 
    counts = collections.Counter(d)
    return np.power(h[0], counts['h']) * np.power(h[1], counts['t'])

print('\nLikelihood of h1 for data_book1:', likelihood(data_book1, h1))
print('Likelihood of h2 for data_book1:', likelihood(data_book1, h2))
print('Likelihood of h3 for data_book1:', likelihood(data_book1, h3))
print('Likelihood of h4 for data_book1:', likelihood(data_book1, h4))
print('Likelihood of h5 for data_book1:', likelihood(data_book1, h5))

print('\nLikelihood of h1 for data_book2:', likelihood(data_book2, h1))
print('Likelihood of h2 for data_book2:', likelihood(data_book2, h2))
print('Likelihood of h3 for data_book2:', likelihood(data_book2, h3))
print('Likelihood of h4 for data_book2:', likelihood(data_book2, h4))
print('Likelihood of h5 for data_book2:', likelihood(data_book2, h5))

print('\nLikelihood of h1 for data_book3:', likelihood(data_book3, h1))
print('Likelihood of h2 for data_book3:', likelihood(data_book3, h2))
print('Likelihood of h3 for data_book3:', likelihood(data_book3, h3))
print('Likelihood of h4 for data_book3:', likelihood(data_book3, h4))
print('Likelihood of h5 for data_book3:', likelihood(data_book3, h5))

print('\nLikelihood of h1 for data_book4:', likelihood(data_book4, h1))
print('Likelihood of h2 for data_book4:', likelihood(data_book4, h2))
print('Likelihood of h3 for data_book4:', likelihood(data_book4, h3))
print('Likelihood of h4 for data_book4:', likelihood(data_book4, h4))
print('Likelihood of h5 for data_book4:', likelihood(data_book4, h5))

print('\nLikelihood of h1 for data_book5:', likelihood(data_book5, h1))
print('Likelihood of h2 for data_book5:', likelihood(data_book5, h2))
print('Likelihood of h3 for data_book5:', likelihood(data_book5, h3))
print('Likelihood of h4 for data_book5:', likelihood(data_book5, h4))
print('Likelihood of h5 for data_book5:', likelihood(data_book5, h5))

posterior = np.zeros(5)
posteriors = np.zeros((11, 5))
posteriors[0] = prior

for n in range(1,11): 
    for (i,h) in enumerate([h1,h2,h3,h4,h5]):
        posterior[i] = (prior[i] * likelihood(data_book1[:n], h))
    posterior /= np.sum(posterior)
    posteriors[n] = posterior

    
x = np.arange(0,11)  
y1 = posteriors[:,0]
y2 = posteriors[:,1]
y3 = posteriors[:,2]
y4 = posteriors[:,3]
y5 = posteriors[:,4]

output_notebook()
p = figure() 
# p.legend.title = 'Bayesian Learning'

for (y,label,color) in zip([y1,y2,y3,y4,y5], 
                     ['P(h1|d)', 'P(h2|d)','P(h3|d)', 'P(h4|d)', 'P(h5|d)'], 
                     ['red','green','blue','purple', 'orange']): 
    p.line(x, y, legend_label=label, line_width=2,color=color)
    p.circle(x,y, color = color)

p.legend.location = "top_left"

show(p)
```

# QUESTION 4 (Expected) - Expectation Maximization - 8 Marks

In this question, you are asked to extend the toy Expectation Maximization algorithm introduced in the learning probabilistic models notebook to handle 3 classes and more than one parameter (each class has its own two parameters: mu and sigma of the Gaussian normal).

Below we generate data for three normally distributed classes of reaction times for the general population, gamers, and pilots. Your task is to:
1. Complete the E step that samples from the pdf of each class and updates the estimates samples for each class
2. Complete the M step that estimates parameters from the samples
3. Once EM has been run, generate 1000 samples from the EM estimated distribution parameters for each class
4. Plot these generated distributions in the same way shown below for the distributions with MLE parameters

```python
import numpy as np
from matplotlib import pyplot

# Gaussian Normal distribution parameters for three classes of reaction time general population, gamers, and Navy fighter pilots
mean_reaction_time_general = 278   # Mean reaction time of the general population is ~250ms
std_reaction_time_general  = 45    # STD of reaction time of the general population is ~30ms
mean_reaction_time_gamers  = 227   # Reaction time of the gamer population is ~227ms
std_reaction_time_gamers   = 33    # STD of reaction time of the general population is ~33ms
mean_reaction_time_pilots  = 223   # Reaction time of the US Navy fighter pilot population is ~223ms
std_reaction_time_pilots   = 5    # STD of reaction time of the US Navy fighter pilot population is ~5ms

# generate 1000 samples of each class 
general_samples = np.random.normal(mean_reaction_time_general, std_reaction_time_general, 1000)
gamer_samples = np.random.normal(mean_reaction_time_gamers, std_reaction_time_gamers, 1000)
pilot_samples = np.random.normal(mean_reaction_time_pilots, std_reaction_time_pilots, 1000)

# Plot this generated data
bins = np.linspace(100, 400, 50)
pyplot.hist(general_samples, bins, alpha=0.5, label='General Population')
pyplot.hist(gamer_samples, bins, alpha=0.5, label='Gamers')
pyplot.hist(pilot_samples, bins, alpha=0.5, label='Pilots')
pyplot.legend(loc='upper right')
pyplot.show()
    
# MLE of MEAN (mu) and STD (sigma) of all classes
mle_mean_general = np.mean(general_samples)
mle_std_general = np.std(general_samples)
mle_mean_gamers = np.mean(gamer_samples)
mle_std_gamers = np.std(gamer_samples)
mle_mean_pilots = np.mean(pilot_samples)
mle_std_pilots = np.std(pilot_samples)

print("MLE Estimates of Normal Distrubiton parameters from generated data")
print("General population mu: ", mle_mean_general, " ; sigma: ", mle_std_general)
print("Gamer population mu: ", mle_mean_gamers, " ; sigma: ", mle_std_gamers)
print("Pilot population mu: ", mle_mean_pilots, " ; sigma: ", mle_std_pilots)

```

```python

from scipy.stats import norm

# First, some initial guess of the parameter we do not know, mu for each class.
em_mean_general = 260 
em_std_general = 30 
em_mean_gamers = 250 
em_std_gamers = 30 
em_mean_pilots = 240 
em_std_pilots = 20

reaction_times = np.hstack([general_samples,gamer_samples,pilot_samples]) # Stack arrays in sequence horizontally (column wise).
np.random.shuffle(reaction_times)

for i in range(0,20): # iterate 20 times, remember EM is an iterative method.
    estimated_general_samples = [] 
    estimated_gamer_samples = [] 
    estimated_pilot_samples = [] 
    print(i," General pop.: ", em_mean_general, em_std_general," Gamers: ", em_mean_gamers, em_std_gamers," Pilots: ", em_mean_pilots, em_std_pilots)

    for r in reaction_times:  
        # E-step - use current model to estimate values for the hidden variable (class membership)
        # Your code goes here
        
        p_general = norm(em_mean_general, em_std_general).pdf(r)
        p_gamers = norm(em_mean_gamers, em_std_gamers).pdf(r)
        p_pilots = norm(em_mean_pilots, em_std_pilots).pdf(r)
        if (p_general > p_gamers and p_general > p_pilots): 
            estimated_general_samples.append(r)
        elif (p_gamers > p_general and p_gamers > p_pilots): 
            estimated_gamer_samples.append(r)
        else:
            estimated_pilot_samples.append(r)
   
    # M-step - using the estimated class values re-calculate the parameters of the model i.e the means 
    # Your code goes here

    em_mean_general = np.mean(estimated_general_samples)
    em_std_general = np.std(estimated_general_samples)
    em_mean_gamers = np.mean(estimated_gamer_samples)
    em_std_gamers = np.std(estimated_gamer_samples)
    em_mean_pilots = np.mean(estimated_pilot_samples)
    em_std_pilots = np.std(estimated_pilot_samples)

# generate 1000 samples of each class 
# Your code goes here
general_samples = np.random.normal(mean_reaction_time_general, std_reaction_time_general, 1000)
gamer_samples = np.random.normal(mean_reaction_time_gamers, std_reaction_time_gamers, 1000)
pilot_samples = np.random.normal(mean_reaction_time_pilots, std_reaction_time_pilots, 1000)

# Plot this generated data
# Your code goes here
bins = np.linspace(100, 400, 50)
pyplot.hist(general_samples, bins, alpha=0.5, label='General Population')
pyplot.hist(gamer_samples, bins, alpha=0.5, label='Gamers')
pyplot.hist(pilot_samples, bins, alpha=0.5, label='Pilots')
pyplot.legend(loc='upper right')
pyplot.show()

print("EM Estimates of Normal Distrubiton parameters from generated data")
print("General population mu: ", em_mean_general, " ; sigma: ", em_std_general)
print("Gamer population mu: ", em_mean_gamers, " ; sigma: ", em_std_gamers)
print("Pilot population mu: ", em_mean_pilots, " ; sigma: ", em_std_pilots)
```

The SciKitLearn toolset also has several datasets for practicing and developing your machine-learning approaches to very different types of data. We are going to consider the wine dataset, details here <https://scikit-learn.org/stable/datasets/toy_dataset.html#wine-recognition-dataset>

Good enough model for ReLu\
( % , % ), with  relu : %\

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_validate
import numpy as np
data = load_wine()

# Your code goes here
# Part 1 simple accuracy function
X = data.data
y = data.target

def simple_accuracy(clf, X, y):
    y_pred = clf.predict(X)
    return np.sum(y_pred == y) / len(y)

layer_sizes = range(1,6)
activations = ['relu', 'tanh', 'logistic']

# YOUR CODE GOES HERE
# Variables for winners
our_accuracy_winner = -1.0
our_layer1_winner = None
our_layer2_winner = None
our_activation_winner = None

cv_accuracy_winner = -1.0
cv_layer1_winner = None
cv_layer2_winner = None
cv_activation_winner = None

# Variable/dict for good enough models
GOOD_ENOUGH_THRESHOLD = 0.95
good_enough_models = {
    'relu': {'accuracy': -1.0, 'l1': None, 'l2': None, 'size': float('inf')},
    'tanh': {'accuracy': -1.0, 'l1': None, 'l2': None, 'size': float('inf')},
    'logistic': {'accuracy': -1.0, 'l1': None, 'l2': None, 'size': float('inf')},
}
for activation in activations:
    for layer1 in layer_sizes:
        for layer2 in layer_sizes:
            # YOUR CODE GOES HERE
            clf = MLPClassifier(hidden_layer_sizes=(layer1, layer2), activation=activation, solver='adam', max_iter=10000,random_state=1,)
            clf.fit(X, y)
            our_acc = simple_accuracy(clf, X, y)
            cv_results = cross_validate(clf, X, y, cv=10, scoring=('accuracy'), error_score=0.0)

            # Calculate mean/std
            test_accuracies = cv_results['test_score']
            cv_mean_acc = np.mean(test_accuracies)
            cv_std_acc = np.std(test_accuracies)

            # Print outputs
            print(f"( {layer1:2d} , {layer2:2d} ), with {activation:10s} our accuracy: {our_acc:.15f} ; cv mean accuracy: {cv_mean_acc:.15f} ; cv accuracy std: {cv_std_acc:.15f}")

            # Check for our accuracy winner
            if our_acc > our_accuracy_winner:
                our_accuracy_winner = our_acc
                our_layer1_winner = layer1
                our_layer2_winner = layer2
                our_activation_winner = activation

            # Check for cross-validation winner
            if cv_mean_acc > cv_accuracy_winner:
                cv_accuracy_winner = cv_mean_acc
                cv_layer1_winner = layer1
                cv_layer2_winner = layer2
                cv_activation_winner = activation

            # Check if good enough
            if cv_mean_acc >= GOOD_ENOUGH_THRESHOLD:
                current_size = layer1 + layer2
                # Check if this is smaller than the current best for this activation
                if current_size < good_enough_models[activation]['size']:
                    good_enough_models[activation]['size'] = current_size
                    good_enough_models[activation]['l1'] = layer1
                    good_enough_models[activation]['l2'] = layer2
                    good_enough_models[activation]['accuracy'] = cv_mean_acc

                # If sizes are equal, prefer the one with higher accuracy
                elif current_size == good_enough_models[activation]['size'] and cv_mean_acc > good_enough_models[activation]['accuracy']:
                     good_enough_models[activation]['l1'] = layer1
                     good_enough_models[activation]['l2'] = layer2
                     good_enough_models[activation]['accuracy'] = cv_mean_acc

# Note the following two lines are examples, you may store these values in your own way
print('Winning model: (', our_layer1_winner, ',', our_layer2_winner, '), with ', our_activation_winner, ' :', our_accuracy_winner)
print('Winning model: (', cv_layer1_winner, ',', cv_layer2_winner, '), with ', cv_activation_winner, ' :', cv_accuracy_winner)

# Example output (... is not the actual output just there are many lines here)
# ( 1 , 1 ), with  relu  our accuracy : 0.398876404494382 ; cv mean accuracy:  0.3993464052287582 ; cv accuracy std: 0.024707265583576204
# ( 1 , 2 ), with  relu  our accuracy : 0.398876404494382 ; cv mean accuracy:  0.3758169934640523 ; cv accuracy std: 0.02926616101599753
# ...
# Winning model: ( 5 , 4 ), with  logistic  : 1.0
# Winning model: ( 5 , 4 ), with  tanh  : 0.961111111111111

print('\nGood enough model for ReLu')
model_relu = good_enough_models['relu']
if model_relu['l1'] is not None:
    print(f'( {model_relu["l1"]} , {model_relu["l2"]} ), with  relu : {model_relu["accuracy"]:.15f}')
else:
    print(f'( % , % ), with  relu : % (None found >= {GOOD_ENOUGH_THRESHOLD})')

print('Good enough model for tanh')
model_tanh = good_enough_models['tanh']
if model_tanh['l1'] is not None:
    print(f'( {model_tanh["l1"]} , {model_tanh["l2"]} ), with  tanh : {model_tanh["accuracy"]:.15f}')
else:
    print(f'( %, % ), with  tanh : % (None found >= {GOOD_ENOUGH_THRESHOLD})')

print('Good enough model for logistic')
model_logistic = good_enough_models['logistic']
if model_logistic['l1'] is not None:
    print(f'( {model_logistic["l1"]} , {model_logistic["l2"]} ), with  logistic : {model_logistic["accuracy"]:.15f}')
else:
    print(f'( % , % ), with  logistic : % (None found >= {GOOD_ENOUGH_THRESHOLD})')
```
